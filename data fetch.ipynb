```python
import requests
from PIL import Image
from io import BytesIO
import os
import pandas as pd
import numpy as np
```

### Saving image file from latitude and longitude

```python
os.makedirs("images", exist_ok=True)

def fetch_satellite_image(lat, lon, idx, zoom=17.5):
    filename = f"images/house_{idx}.png"
    if os.path.exists(filename):
        return
    url = (
        f"https://api.mapbox.com/styles/v1/mapbox/satellite-v9/static/"
        f"{lon},{lat},{zoom}/224x224"
        f"?access_token={MAPBOX_TOKEN}"
    )
    
    response = requests.get(url)
    image = Image.open(BytesIO(response.content))
    image.save(f"images/house_{idx}.png")
```
